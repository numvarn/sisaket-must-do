# มาศรีสะเกษ ต้องมาค้นหาหรือมาทำอะไร (Must Seek)

```python
must_seek = df[["must_seek_1", "must_seek_2", "must_seek_3","must_seek_4"]]
```

```python
must_seek.head()
```

### สิ่งแรก ที่ผู้ตอบคิดถึงเกี่ยวกับ มาศรีสะเกษ ต้องมาค้นหาหรือมาทำอะไร (Must Seek)

```python
first_choice = must_seek["must_seek_1"]

results = first_choice.value_counts(dropna=True).head(5)

results.to_frame('counts')
```

### ความคิดที่ 2 : ที่ผู้ตอบคิดถึงเกี่ยวกับ มาศรีสะเกษ ต้องมาค้นหาหรือมาทำอะไร (Must Seek)

```python
second_choice = must_seek["must_seek_2"]

results = second_choice.value_counts(dropna=True).head(5)

results.to_frame('counts')
```

### ผลของการนับความถี่ของคำตอบ ทั้งหมด : ในประเด็น ที่ผู้ตอบคิดถึงเกี่ยวกับ มาศรีสะเกษ ต้องมาค้นหาหรือมาทำอะไร (Must Seek)

```python
# รวมข้อมูลจากทั้งสี่คอลัมน์ให้เป็นคอลัมน์เดียว
all_must_seek = df[['must_seek_1', 'must_seek_2', 'must_seek_3', 'must_seek_4']].melt(value_name="must_seek")

# นับความถี่ของค่าที่รวมแล้ว
frequency_all = all_must_seek['must_seek'].value_counts(dropna=True)

# แสดงผล
frequency_all.to_frame('counts').head(5)
```
